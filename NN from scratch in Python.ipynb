```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[['age','affordibility']], df['bought_insurance'], test_size=0.2, random_state=25)
```

```python
X_train_scaled=X_train.copy()
X_train_scaled=X_train_scaled/100
X_test_scaled=X_test.copy()
X_test_scaled=X_test_scaled/100
```

```python
import numpy as np
```

```python
def sigmoid_numpy(x):
    return 1/(1+np.exp(-x))
```

```python
def log_loss(y_true, y_pred):
    epsilon= 1e-15
    y_pred_new=[max(i,epsilon) for i in y_pred]
    y_pred_new=[min(i,1-epsilon) for i in y_pred_new]
    y_pred_new = np.array(y_pred_new)
    return np.mean(y_true*np.log(y_pred_new)+(1-y_true)*np.log(1-y_pred_new))
```

```python
def grad_desc(age,afford,y_true,epochs):
    w1=w2=1
    bias=0
    rate=0.5
    n=len(afford)
    for i in range(epochs):
        weighted_sum=w1*age+w2*afford+bias
        y_pred=sigmoid_numpy(weighted_sum)
        loss=log_loss(y_true,y_pred)
        
        w1_d=(1/n)*np.dot(np.transpose(age),(y_pred-y_true))
        w2_d=(1/n)*np.dot(np.transpose(afford),(y_pred-y_true))
        bias_d=np.mean(y_pred-y_true)
        
        w1=w1-rate*w1_d
        w2=w2-rate*w2_d
        bias=bias-rate*bias_d
        
        
        print(f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
        #if loss<=loss_thres:
            #break
    return w1,w2,bias
```

```python
grad_desc(X_train_scaled['age'], X_train_scaled['affordibility'], y_train, 1000)
```
